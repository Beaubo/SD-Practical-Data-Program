This is my work through a basic pandas tutorial, which covers basic mechanics and manipulation of Python data frames, and basic statistical operations.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 100)

print 'Pandas version ' + pd.__version__
print 'Numpy version ' + np.__version__
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/apawlik/2014-04-09-GARNET/gh-pages/lessons/garnet/pandas/data/weather_year.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
#Summary statistics
data.describe()
```

```python
#Accessing columns: We can do it like a python dictionary, with the "key" (column name)
data["EDT"]
```

```python
#We can get multiple rows by using a list of string keys (thus, the double brackets)
data[["EDT", "Mean TemperatureF"]]
```

```python
#We can also use "dot syntax" to access columns
data.EDT
```

```python
#How many unique values are there in " CloudCover"
data[' CloudCover'].unique()
```

```python
data[' CloudCover'].value_counts()
```

```python
#Plot the unique values for cloud cover and their frequency in a bar graph
data[' CloudCover'].value_counts().plot(kind='bar')
```

```python
#Summary stats (mean, SD, quartiles) for variable)
data["Mean TemperatureF"].describe()
```

```python
data["Mean TemperatureF"].mean()
```

```python
data["Mean TemperatureF"].std()
```

```python
#Mean of every variable in dataset
data.mean()
```

```python
#Quick Plots
data["Mean TemperatureF"].plot()
```

```python
data["Mean TemperatureF"].hist(bins=20)
```

```python
#Selecting rows of interest from data frame, making another data frame

freezing_days = data[data['Max TemperatureF'] <= 32]
```

```python
freezing_days.head()
```

```python
#Messing with the data frame

pretty_cold = data[(data['Max TemperatureF'] <= 32) & (data['Max TemperatureF'] >= 18)] 
type(pretty_cold)
```

```python
##Group by lets us group our rows by the value of a certain column (in this case, value of cloud cover)
#and then aggregates the mean over those groups

data.groupby([' CloudCover']).mean()
```

```python
data.head()
```

```python
#Converting date variable to date-time type, and indexing the data frame by the type 

data.EDT = pd.to_datetime(data.EDT)
data.index = data.EDT
```
